```python
from pyspark import SparkContext
sc = SparkContext (
    master = "local",
    appName = "Strm1"
)
```

```python
rdd = sc.textFile("../Udemy_Python/Spark Streaming/book.txt")
```

```python
words = rdd.flatMap(lambda word: word.split(' '))
```

```python
lowerWords = words.map(lambda word: word.lower())
```

```python
wordCount = lowerWords.countByValue()
```
